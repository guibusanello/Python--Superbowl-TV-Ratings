```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('super-bowl-ratings.csv')
data
```

```python
data.head()
```

```python
# Entendendo os dados
print(data.shape)
print(data.columns)
```

```python
# Entendendo os dados
data.info()
```

```python
# Resumo estatístico das variáveis numéricas
data = data.drop('super_bowl_number', axis=1)
pd.options.display.float_format = '{:.2f}'.format
data.describe()
```

```python
# Resumo das variáveis categóricas
data.network.value_counts().unique
```

```python
# Gráfico de barras com a contagem do número de transmissões do SB por emissora

plt.figure(figsize=(8,4))
plt.title('Network')
colors = sns.color_palette("hsv", len(data['network'].unique()))
sns.countplot(x=data['network'], palette=colors)
```

```python
# Converter a coluna 'date' para datetime
data['date'] = pd.to_datetime(data['date'])
data.head()
```

```python
# Criar uma nova coluna 'year' com somente o ano
data['year'] = data['date'].dt.year
data.head()
```

```python
# Plotando valor médio de espectadores
average_viewers_mean = data['average_viewers'].mean()
average_viewers_mean
```

```python
# Gráfico de linhas mostrando a evolução da média de espectadores do SB

sns.set_style("darkgrid")
plt.figure(figsize=(8,5))
plt.title("Superbowl Average Viewers")
plt.ylabel("average_viewers",size=10)
plt.xlabel("year",size=10)
plt.xticks(rotation=90)
sns.lineplot(data=data, x='year', y ='average_viewers',marker='o', color='blue')
plt.axhline(y=average_viewers_mean, color='red', linestyle='--', label=f'Mean ({average_viewers_mean:.2f})')
```

```python
# Plotando valor médio do custo da propaganda
average_adcost_mean = data['cost_of_30_second_ad_usd'].mean()
average_adcost_mean
```

```python
# Gráfico de linhas mostrando a evolução do custo por 30 segundos de propaganda na transmissão do SB

sns.set_style("darkgrid")
plt.figure(figsize=(8,5))
plt.title("Cost of 30 second ad usd")
plt.ylabel("cost_of_30_second_ad_usd",size=10)
plt.xlabel("year",size=10)
plt.xticks(rotation=90)
sns.lineplot(data=data, x='year', y ='cost_of_30_second_ad_usd',marker='o', color='green')
plt.axhline(y=average_adcost_mean, color='blue', linestyle='--', label=f'Mean ({average_adcost_mean:.2f})')
```

```python
# Household Rating ao longo do tempo

plt.figure(figsize=(10,3))
plt.title("Household Rating")
plt.xticks(rotation=90)
palette = sns.color_palette("viridis", len(data))
sns.barplot(x=data["year"], y=data["household_rating"], palette=palette)
```

```python
# Household share

plt.figure(figsize=(10,3))
plt.title("Household Share")
plt.xticks(rotation=90)
palette = sns.color_palette("viridis", len(data))
sns.barplot(x=data["year"], y=data["household_share"], palette=palette)
```

```python
# Verificando a correlação entre as variáveis numéricas

numeric_cols = data.select_dtypes(include=['number']).columns
correlation_matrix = data[numeric_cols].corr()
correlation_matrix
```

```python
# Matriz de correlação

plt.figure(figsize=(8,4))
sns.heatmap(correlation_matrix, cmap='CMRmap', annot=True)
plt.show()
```
